```python
# prompt: 로봇이 좌표에 따라 움직이는 게임을 만들어줘. 첫번째로 위로 한칸, 두번째로 오른쪽으로 한칸, 세번째로 아래로 한칸, 마지막으로 왼쪽으로 한칸 움직이는 코드를 작성해줘. 이걸 그림으로 표현해줘. 좌표를 움직일 때마다 그림을 새로 그려줘.

# Initial position of the robot
robot_x = 0
robot_y = 0

# Store the history of robot positions for drawing
position_history = [(robot_x, robot_y)]

def draw_robot_movement(history):
    plt.figure()
    plt.title("로봇 움직임")
    plt.axis('equal')
    plt.grid(True)

    # Extract x and y coordinates from history
    x_coords = [pos[0] for pos in history]
    y_coords = [pos[1] for pos in history]

    # Determine plot limits dynamically based on history
    if history:
        x_min, x_max = min(x_coords), max(x_coords)
        y_min, y_max = min(y_coords), max(y_coords)
        padding = 2 # Add some padding around the path
        plt.xlim(x_min - padding, x_max + padding)
        plt.ylim(y_min - padding, y_max + padding)

    # Draw the path
    plt.plot(x_coords, y_coords, 'bo-') # Blue circles for positions, line connecting them

    # Draw the current robot position (last point in history)
    plt.plot(x_coords[-1], y_coords[-1], 'ro', markersize=10) # Red circle for current position

    plt.show()

# Draw the initial position
draw_robot_movement(position_history)
time.sleep(1) # Pause for 1 second

# Move 1: Up (Increase y)
robot_y += 1
position_history.append((robot_x, robot_y))
draw_robot_movement(position_history)
time.sleep(1) # Pause for 1 second

# Move 2: Right (Increase x)
robot_x += 1
position_history.append((robot_x, robot_y))
draw_robot_movement(position_history)
time.sleep(1) # Pause for 1 second

# Move 3: Down (Decrease y)
robot_y -= 1
position_history.append((robot_x, robot_y))
draw_robot_movement(position_history)
time.sleep(1) # Pause for 1 second

# Move 4: Left (Decrease x)
robot_x -= 1
position_history.append((robot_x, robot_y))
draw_robot_movement(position_history)
time.sleep(1) # Pause for 1 second
```
